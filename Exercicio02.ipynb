```python
import matplotlib.pyplot as plt
import random
```

```python
# Função objetivo
def f(x):
  return -x**2 + 4*x

# Definindo o intervalo
intervalo_min = 0
intervalo_max = 4

# Definindo o tamanho da vizinhança
delta = 0.01

# Definindo o número máximo de iterações
max_iter = 100

# Lista para armazenar os pontos explorados
pontos_explorados = []

# Inicialização
x = random.uniform(intervalo_min, intervalo_max)
pontos_explorados.append(x)

# Loop principal
for _ in range(max_iter):
  # Avaliação da vizinhança
  vizinho_esquerda = x - delta
  vizinho_direita = x + delta
  f_vizinho_esquerda = f(vizinho_esquerda)
  f_vizinho_direita = f(vizinho_direita)

  # Atualização da solução
  if f_vizinho_esquerda > f(x):
    x = vizinho_esquerda
  elif f_vizinho_direita > f(x):
    x = vizinho_direita
  else:
    # Parada: não há vizinhos com valor maior
    break

  # Armazenando o ponto explorado
  pontos_explorados.append(x)

# Imprimindo o resultado
print(f"Máximo local encontrado: x = {x}")
print(f"Valor máximo: f(x) = {f(x)}")

# Adicionando a cor aos pontos
cores = [f(x) for x in pontos_explorados]

# Plotando o gráfico de dispersão
plt.scatter(pontos_explorados, cores, c=cores, cmap="viridis")
plt.xlabel("Iteração")
plt.ylabel("Valor de x")
plt.colorbar()
plt.show()
```
